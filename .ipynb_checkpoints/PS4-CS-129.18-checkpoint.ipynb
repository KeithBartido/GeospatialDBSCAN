```python
import numpy as np
from sklearn.cluster import KMeans
import folium
from folium import plugins
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import foursquare

plt.style.use('ggplot')

pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
df = pd.read_csv("coffeeshops.csv")
df = df.drop_duplicates()
```

```python
df.head()
```

### 1. Top 3 Coffeeshops

```python
df['name'].value_counts().head(3)
```

### 2.
The latitude variance (0.0071) and longitude variance (0.0015) are different because latitude is of a y-axis while longitude is of a x-axis. The fact that latitude variance is higher suggests that the coffeeshops are more spread out from each other on a north-south direction (suggesting that coffeeshops are spread throughout the islands to the north and south, not just Metro Manila)
The coffeeshops are closer to one another, even if by a slight margin, in an east-west direction. 

```python
lat_var = np.var(df["lat"])
lng_var = np.var(df["lng"])
print(f"Latitude variance: {lat_var}")
print(f"Longitude variance: {lng_var}")
```

```python
X = np.array(df[["lat", "lng"]].values.tolist())
plt.scatter(X[:, 0], X[:, 1], s=50)
```

### 3. 
KMeans inertia is the "sum of squared distances of samples to their closest cluster center." In general, it measures how far away the points within a cluster are.   
On the other hand, DBSCAN Epsilon is an inputted parameter considered in grouping points, specifying how close points should be to each other to be considered a part of a cluster.

Contrary to inertia that **is stated** after KMeans (output), DBSCAN Epsilon is **to be stated** as an input parameter.

> *Sources:  
https://github.com/alitouka/spark_dbscan/wiki/Choosing-parameters-of-DBSCAN-algorithm    
https://towardsdatascience.com/k-means-clustering-from-a-to-z-f6242a314e9a*

```python
kmeans = KMeans(n_clusters=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```python
df["cluster"] = y_kmeans
```

```python
palette = sns.color_palette("pastel", 10).as_hex()
palette
```

```python
color_getter = lambda x: palette[int(x)]
df["color"] = df["cluster"].map(color_getter)
df.head()
```

```python
basemap = folium.Map(location=[14.5562297,121.0204097], tiles="cartodbdark_matter", zoom_start=15)

for index, row in df.iterrows():
    folium.CircleMarker([row["lat"], row["lng"]], radius=5,
        fill=True,
        color =row["color"],
        popup = folium.Popup(row["name"], parse_html=True),
        fill_color=row["color"]).add_to(basemap)
    
basemap.save('coffeeshops.html')

basemap
```

```python
sum_of_squared_distances = []
K = range(1,30)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
xp=df.groupby('name').count().sort_values(by ='id', ascending=False).head(3)
xp
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.005, min_samples=5).fit(X)
y_dbscan = dbscan.labels_
```

```python
y_dbscan
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, s=50, cmap='viridis')
```

```python
df["dbscan_cluster"] = y_dbscan
palette = sns.color_palette("pastel", len(y_dbscan)).as_hex()
color_getter = lambda x: palette[int(x)]
df["db_color"] = df["dbscan_cluster"].map(color_getter)
df.head()
```

### 4. Clustered locations w/ DBSCAN

```python
basemap = folium.Map(location=[14.5562297,121.0204097], tiles="cartodbdark_matter", zoom_start=12)

for index, row in df.iterrows():
    folium.CircleMarker([row["lat"], row["lng"]], radius=5,
    fill=True,
    color =row["db_color"],
                        popup = folium.Popup(row["name"], parse_html=True),
    fill_color=row["db_color"]).add_to(basemap)
    
basemap.save('coffeeshops.html')

basemap
```

### 5.

- The clusters show areas where there is a high density of coffeeshops close to one another.
- The # of clusters, 10, shows that there are more than a handful of areas (both big and small areas) with plenty coffeeshops. The single dominant cluster that seems to spread across Metro Manila suggests that Metro Manila in itself has a multitude of coffeeshops.
- This data shows the concentration of coffeeshops in the country. It could help us infer that the high density clusters of coffeeshops could possibly be in response to, and in accompaniment with, high density of urban activity in those areas. It is also absolutely useful to take comfort in knowing that you won't run out of coffeeshops if you so happen to be in those clustered locations.
